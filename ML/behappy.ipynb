```python
import pandas as pd
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
a = pd.read_csv("toyota_model.csv")
```

```python
a = a.drop(["Unnamed: 0", "YM", "Brand" ], 1)
```

```python
models = pd.get_dummies(a['Model'])
```

```python
age_sex = pd.get_dummies(a["Target_Audience"])
```

```python
a = a.drop(["Model", "Target_Audience"], 1)
```

```python
y = a['count_reg']
```

```python
a = a.drop(["count_reg"], 1)
```

```python
stat = a.describe().T
```

```python
data = pd.concat([a, age_sex, models], 1)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,y, test_size=0.7, random_state=13)
```

```python
stat = X_train.describe().T

X_train = (X_train - stat["mean"])/stat["std"]
X_test = (X_test - stat["mean"])/stat["std"]
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
```

```python
model = keras.Sequential([

layers.Dense(64, activation='relu', input_shape=[len(data.keys())]),  
layers.Dense(64, activation='relu' ),  

layers.Dense(1)
])

optimizer = tf.keras.optimizers.Adam(learning_rate=0.00005)

model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
model.summary()
```

```python
history = model.fit(
X_train, y_train, batch_size=3,
epochs=3000 , verbose=1, validation_data=(X_test, y_test))
```

```python
model.save("happy.h5")
```
