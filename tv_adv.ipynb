```python
df_t = df.drop(index=df[(df['Label_bsq'] == 0) | (df['Label_bsq_1m'] == 0) | (df['Label_bsq_2m'] == 0) | (df['Label_bsq_3m'] == 0)].index)
```

```python
df_t = df_t.drop(index=df_t[(df_t['Label_ym_nr'] == 0) | (df_t['Label_ym1_nr'] == 0) | (df_t['Label_ym2_nr'] == 0) | (df_t['Label_ym3_nr'] == 0) | (df_t['Label_ym4_nr'] == 0) | (df_t['Label_ym5_nr'] == 0)].index)
```

```python
df_t.drop(columns=df_t.columns[:19], axis=1, inplace=True)
```

```python
df_t.columns[-10:]
```

```python
for i in df_t.columns[-10:]:
    X = df_t.drop(columns=df_t.columns[-10:], axis=1)
    y = np.log10(df_t[i])

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)

    predicted = rf.predict(X_test)

    fi = pd.Series(rf.feature_importances_, index=X.columns)
    print('\033[91m' + '\033[1m' + i + '\033[0m')
    print('mean_squared_errors: {}'.format(mean_squared_error(y_test, predicted)))
    print('r2_score: {}'.format(r2_score(y_test, predicted)))
    print(fi.sort_values(ascending=False))
```

# 브랜드 : 광고 - 검색

```python
df[(df['Label_bsq'] == 0) | (df['Label_bsq_1m'] == 0) | (df['Label_bsq_2m'] == 0) | (df['Label_bsq_3m'] == 0)].index
```

```python
df_n = df.drop(index=df[(df['Label_bsq'] == 0) | (df['Label_bsq_1m'] == 0) | (df['Label_bsq_2m'] == 0) | (df['Label_bsq_3m'] == 0)].index)
```

```python
for i in df_n['Brand'].unique():

    df_n[df_n['Brand'] == i].columns[:19]

    data = df_n[df_n['Brand'] == i].drop(columns=df_n[df_n['Brand'] == i].columns[:19], axis=1)

    X = data.drop(columns=data.columns[-10:], axis=1)
    y = np.log10(data['Label_bsq_3m'])

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
    
    rf = RandomForestRegressor()
    rf.fit(X_train, y_train)

    predicted = rf.predict(X_test)

    actual = y_test

    act_df = pd.DataFrame(actual)
    pred_df = pd.DataFrame(predicted)

    act_df.reset_index(inplace=True)

    act_pred_df = pd.concat([act_df, pred_df], axis=1)

    act_pred_df.columns = ['index','actual', 'predicted']

    plt.figure(figsize=(15,10))
    plt.plot(act_pred_df['predicted'],act_pred_df['actual'], 'o')

    plt.xlabel('predicted', fontsize=14)

    plt.ylabel('actual', fontsize=14)

    plt.axis([0, 10, 0, 10])  

    line = plt.plot([0,30], [0,30])
    plt.setp(line, color='r', linewidth=3.0)
    plt.title(i, fontsize=30, y=1.05)

    print(plt.show())

    fi = pd.Series(rf.feature_importances_, index=X.columns)
    print(i)
    print('mean_squared_errors: {}'.format(mean_squared_error(y_test, predicted)))
    print('r2_score: {}'.format(r2_score(y_test, predicted)))
    print(fi.sort_values(ascending=False))
```
